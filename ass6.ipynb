Q1. Ask the user to enter a value (e.g., M) for a mathematical function.
 Generate x values from -10 to 10 using np.linspace().
 Compute y values for:
o y= M⋅x
2
o y=M⋅sin(x)
 Plot both functions in a single figure:
o Use different colors and line styles.
o Add legend, grid, and title.

```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
M = float(input("Enter a value for M in the mathematical function: "))
x = np.linspace(-10,10,100)
y1 = M * x**2
y2 = M * np.sin(x)

plt.plot(x, y1, 'r--', label = '$y1 = M * x^2$')
plt.plot(x, y2, 'b-', label = '$y2 = M * sin(x)$')
plt.title('Plot of Mathematical Functions')
plt.legend()
plt.xlabel("x values ---->")
plt.ylabel("y values ---->")
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```

Q.2 Create a dataset of five subjects and scores.
 Convert it into a Pandas DataFrame.
 Plot the scores using a Seaborn bar plot with:
o Different colors for each bar.
o Annotations on top of each bar.
o Title, axis labels, and grid.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Creating the dataset
myDataset = {
    'Subjects': ['Maths', 'Physics', 'Chemistry', 'Biology', 'Computer Science'],
    'Marks': [98.5, 95, 96, 97.4, 90.5]
}

# Converting to DataFrame
df = pd.DataFrame(myDataset)

# Setting Seaborn style
sns.set_style("whitegrid")

# Creating the bar plot with different colors
#colors = ['violet', 'skyblue', 'magenta', 'seagreen', 'yellow']
ax = sns.barplot(x = 'Subjects', y = 'Marks', data = df, hue = 'Subjects', palette = 'rainbow' , legend = False)

# Adding annotations on top of each bar
for i, mark in enumerate(df['Marks']):
    ax.text(i, mark + 0.5, f"{mark}", ha = 'center', fontsize = 12, fontweight = 'bold')

# Adding labels, title, and grid
plt.title("Scores in Different Subjects", fontsize=14)
plt.xlabel("Subjects", fontsize = 12)
plt.ylabel("Marks", fontsize = 12)
plt.xticks(rotation = 15)
plt.grid(axis='y', linestyle='--', alpha = 0.7)

# Show the plot
plt.show()
```

Q3. Write a Python script to select your roll number as seed for NumPy and
then generate a dataset of 50 values using np.random.randn().
 Create a 2x2 subplot layout displaying:
o Line plot showing cumulative sum.
o Scatter plot with random noise.
 Customize with titles, axis labels, and grids.

```python
roll_number = int(input("Enter your roll number as seed: "))
np.random.seed(roll_number)

# Generate 50 random values from a normal distribution
data = np.random.randn(50)

# Compute cumulative sum
cumsum_data = np.cumsum(data)

# Create a figure with a 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Line plot for cumulative sum
axes[0, 0].plot(cumsum_data, linestyle='-', color='b')
axes[0, 0].set_title("Cumulative Sum Line Plot")
axes[0, 0].set_xlabel("Index")
axes[0, 0].set_ylabel("Cumulative Sum")
axes[0, 0].grid(True)

# Scatter plot with random noise
axes[0, 1].scatter(range(50), data, color='r', alpha=0.7)
axes[0, 1].set_title("Scatter Plot with Random Noise")
axes[0, 1].set_xlabel("Index")
axes[0, 1].set_ylabel("Random Values")
axes[0, 1].grid(True)

# Hide the empty subplots (since we need only 2 plots)
axes[1, 0].axis('off')
axes[1, 1].axis('off')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()
```

Q.4 Download Data-set from the below link
https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv
Apply ‘seaborn’ library to do the following.
1. Read Total profit of all months and show it using a line plot.
2. Read all product sales data and show it using a multiline plot.
3. Plot bar chart for all the features/attributes.

```python
dataset = pd.read_csv('company_sales_data.csv')
dataset.head()
```

```python
# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Create a line plot for Total Profit per month
plt.figure(figsize=(10, 6))
sns.lineplot(data=dataset, x='month_number', y='total_profit', marker='o', color='b')

# Add titles and labels
plt.title('Total Profit per Month')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.xticks(dataset['month_number'])  # Ensure all months are shown
plt.grid(True)
plt.show()
```

```python
# Melt the dataframe to have a long-form representation
product_columns = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
data_melted = dataset.melt(id_vars = ['month_number'], value_vars = product_columns, var_name = 'Product', value_name = 'Sales')

# Create a multiline plot for product sales data
plt.figure(figsize = (12, 8))
sns.lineplot(data = data_melted, x = 'month_number', y = 'Sales', hue = 'Product', marker = 'o')

# Add titles and labels
plt.title('Monthly Sales of Different Products')
plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.xticks(dataset['month_number'])  # Ensure all months are shown
plt.legend(title='Product')
plt.grid(True)
plt.show()
```

```python
# Calculate the total sales for each product
total_sales = dataset[product_columns].sum().reset_index()
total_sales.columns = ['Product', 'Total Sales']
print(total_sales)
print()

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(data = total_sales, x = 'Product', y = 'Total Sales', hue = 'Product', palette = 'rainbow')
plt.title('Total Sales of Each Product')
plt.xlabel('Product')
plt.ylabel('Total Sales Units')
plt.grid(True, axis = 'y')
plt.show()
```
